```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(clinical_trial_data, mouse_data, how='left', on=['Mouse ID','Mouse ID'])

# Display the data table for preview
combined_data.head(20)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume_data = combined_data.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']

# Convert to DataFrame
mean_tumor_volume_data_df = pd.DataFrame(mean_tumor_volume_data)

# Preview DataFrame first reset index to make it match the output
mean_tumor_volume_data_df = mean_tumor_volume_data_df.reset_index()
```

```python
mean_tumor_volume_data_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stderror_tumor_volume = combined_data.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
stderror_tumor_volume_df = pd.DataFrame(stderror_tumor_volume)

# Preview DataFrame
stderror_tumor_volume_df = stderror_tumor_volume_df.reset_index()
```

```python
stderror_tumor_volume_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_volume_data_pivot_df = mean_tumor_volume_data_df.pivot(index = 'Timepoint', columns = 'Drug')['Tumor Volume (mm3)']
# Preview that Reformatting worked
mean_tumor_volume_data_pivot_df
```

```python
# Minor Data Munging to Re-Format the Data Frames Std error
stderror_tumor_volume_pivot_df = stderror_tumor_volume_df.pivot(index = 'Timepoint', columns = 'Drug')['Tumor Volume (mm3)']
# Preview that Reformatting worked
stderror_tumor_volume_pivot_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_tumor_volume_data_pivot_df.index, mean_tumor_volume_data_pivot_df['Capomulin'],
             yerr= stderror_tumor_volume_pivot_df['Capomulin'], color = 'r', marker = 'o', markersize = 7)
plt.errorbar(mean_tumor_volume_data_pivot_df.index, mean_tumor_volume_data_pivot_df['Infubinol'],
             yerr= stderror_tumor_volume_pivot_df['Infubinol'], color = 'b', marker = "^", markersize = 7)

plt.errorbar(mean_tumor_volume_data_pivot_df.index, mean_tumor_volume_data_pivot_df['Ketapril'],
             yerr= stderror_tumor_volume_pivot_df['Ketapril'], color = 'g', marker = "s", markersize = 7)
plt.errorbar(mean_tumor_volume_data_pivot_df.index, mean_tumor_volume_data_pivot_df['Placebo'],
             yerr= stderror_tumor_volume_pivot_df['Placebo'], color = 'black', marker = "D", markersize = 7)


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend([ 'Capomulin','Infubinol','Ketapril','Placebo' ])
plt.grid(axis='y')
plt.show()

# Save the Figure
plt.savefig("tumorResponsePlot.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_site_mean = combined_data.groupby(['Drug','Timepoint']).mean()['Metastatic Sites']
# Convert to DataFrame
meta_site_mean_df = pd.DataFrame(meta_site_mean)
# Preview DataFrame
meta_site_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_site_std_error = combined_data.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
meta_site_std_error_df = pd.DataFrame(meta_site_std_error)
# Preview DataFrame
meta_site_std_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_site_mean_df = meta_site_mean_df.reset_index()
meta_site_mean_pivot_df = meta_site_mean_df.pivot(index = 'Timepoint', columns='Drug')['Metastatic Sites']

# Preview that Reformatting worked
meta_site_mean_pivot_df
```

```python
meta_site_std_error_df = meta_site_std_error_df.reset_index()
meta_site_std_error_pivot_df = meta_site_std_error_df.pivot(index = "Timepoint", columns = "Drug")['Metastatic Sites']
meta_site_std_error_pivot_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(meta_site_mean_pivot_df.index, meta_site_mean_pivot_df['Capomulin'], 
             yerr=meta_site_std_error_pivot_df['Capomulin'], color = 'r', marker = 'o', markersize = 4, linewidth=0.5)
plt.errorbar(meta_site_mean_pivot_df.index, meta_site_mean_pivot_df['Infubinol'], 
             yerr=meta_site_std_error_pivot_df['Infubinol'], color = 'b', marker = '^', markersize = 4, linewidth=0.5)
plt.errorbar(meta_site_mean_pivot_df.index, meta_site_mean_pivot_df['Ketapril'], 
             yerr=meta_site_std_error_pivot_df['Ketapril'], color = 'g', marker = 's', markersize = 4, linewidth=0.5)
plt.errorbar(meta_site_mean_pivot_df.index, meta_site_mean_pivot_df['Placebo'], 
             yerr=meta_site_std_error_pivot_df['Placebo'], color = 'black', marker = 'D', markersize = 4, linewidth=0.5)

plt.legend(['Capomulin','Infubinol','Ketaprilsss','Placebo'])
plt.grid(axis='y')
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")

# Save the Figure
plt.savefig("MetastaticSpread.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Generate the Plot (Accounting for percentages)
plt.plot((mouse_count_df_pivot["Capomulin"] / 25) * 100  )
plt.plot((mouse_count_df_pivot["Infubinol"] / 25) * 100  )
plt.plot((mouse_count_df_pivot["Ketapril"] / 25) * 100  )
plt.plot((mouse_count_df_pivot["Placebo"] / 25) * 100  )

# Save the Figure

# Show the Figure
plt.show()
```
